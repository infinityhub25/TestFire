```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset=pd.read_csv("Algerian_forest_fires_dataset.csv")
```

```python
dataset.head()
```

```python
dataset.info()
```

## missing values

```python
dataset[dataset.isnull().any(axis=1)]
```

```python
# till bejaija region set index=0 
# till sidi-bel abbes region index=1
```

```python
dataset.loc[:122, 'Region']=0
dataset.loc[122:, 'Region']=1
df=dataset
```

```python
df.info()
```

```python
df[['Region']]=df[['Region']].astype(int)
```

```python
df.head()
```

```python
## Remove the null values
df=df.dropna().reset_index(drop=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df=df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
# observe that there is spaces in col name so to fix it
```

```python
df.columns=df.columns.str.strip()
df.columns
```

```python
df.info()
```

```python
## all object type but it should be numerical or float acc to dataset
```

#change the object as integer

```python
df[['month','day','year','Temperature','RH','Ws']]=df[['month','day','year','Temperature','RH','Ws']].astype(int)
```

```python
df.info()
```

```python
##changing the other colms to float datatype
```

```python
objects=[features for features in df.columns if df[features].dtypes=='O']
```

```python
for i in objects:
    if i!='Classes':
        df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
df.to_csv('Algerian_forest_fires_cleaned_dataset.csv', index=False)
```

## after cleaning we are going to start with EDA

```python
##drop day month and year
df_copy=df.drop(['day','month','year'],axis=1)
```

```python
df_copy.head()
```

```python
df_copy['Classes'].value_counts()
df_copy.head()
```

```python
df_copy['Classes']=np.where(df_copy['Classes'].str.contains('not fire'),0,1)
df_copy.head()
```

```python
df_copy.tail()
```

```python
#Categories in class
df_copy['Classes'].value_counts()
```

```python
## Plot density plot for all features
plt.style.use('seaborn-v0_8')
df_copy.hist(bins=50,figsize=(20,15))
plt.show()
```

```python
## Percentage for pieChart
percentage=df_copy['Classes'].value_counts(normalize=True)*100
```

```python
## plotting piechart
classlabels=["Fire","Not Fire"]
plt.figure(figsize=(12,7))
plt.pie(percentage,labels=classlabels,autopct='%1.1f%%')
plt.title("Pie Chart of Classes")
plt.show()
```
